```python
import requests
from bs4 import BeautifulSoup

# Make the HTTP request
url = 'https://www.international.gc.ca/world-monde/international_relations-relations_internationales/sanctions/consolidated-consolide.aspx?lang=eng'
response = requests.get(url)

# Parse the HTML content with Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

# Remove unwanted elements (script, style, comments, etc.)
for element in soup(['script', 'style', 'comment']):
    element.extract()

# Extract the visible text
text = soup.get_text(separator=' ')

# Print the extracted text
print(text)
```

```python
import requests
from bs4 import BeautifulSoup

# Make the HTTP request
url = 'https://www.international.gc.ca/world-monde/international_relations-relations_internationales/sanctions/consolidated-consolide.aspx?lang=eng'
response = requests.get(url)

# Parse the HTML content with Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

# Remove unwanted elements (script, style, comments, etc.)
for element in soup(['script', 'style', 'comment']):
    element.extract()

    # Extract the visible text
text = soup.get_text(separator=' ')

# Accept user input for the name to search
search_name = input("Enter a name to search: ")

# Check if the name exists in the extracted text
if search_name in text:
    print(f"{search_name} is present in the scraped data.")
else:
    print(f"{search_name} is not found in the scraped data.")
```
